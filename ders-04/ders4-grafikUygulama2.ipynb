```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
DF = pd.read_csv("./ds_salaries.csv")
maas = DF["salary"].head(7)
tecrube = DF["job_title"].head(7)
tecrube
```

```python
veriSeti = np.random.randint(10,30,6)
basliklar = ["emre", "eren", "arda", "EMRE", "EREN", "ARDA","asdf"]

figure, eksen = plt.subplots() #alt grafik oluşturur.
eksen.pie(maas, labels=tecrube, autopct ="%1.1f%%")
plt.title("kişi maas")
plt.show()

```

```python
maas2 = DF["salary"].head(10)
maasUSD = DF["salary_in_usd"].head(10)
meslek = DF["job_title"].head(10)
```

```python
fig, ax =plt.subplots() 
ax.barh(meslek, maas2)
ax.set_title("Maas - Meslek Grafigi")
ax.legend(title = "Maaslar")
plt.show()

fig2 , ax2 = plt.subplots()
ax2.barh(meslek, maasUSD)
ax2.set_title("Meslek - USD Maas Grafigi")
ax2.legend(title = "USD Maaslar")
plt.show()
```
