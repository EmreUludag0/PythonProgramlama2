```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
yas = np.array([10,20,30,40,50])
kilo = np.array([15,90,34,80,50])

yas, kilo
```

```python
plt.plot(yas, kilo, "go--")
plt.xlabel("yaslar")
plt.ylabel("kilolar")
```

```python
dizi = np.linspace(0,20,10)
carpim = dizi ** 2
rnd1 = np.random.randint(50,300,10)
rnd2 = np.random.randint(50,300,10)

plt.subplot(1,4,1)
plt.plot(dizi,carpim, "b--")

plt.subplot(1,4,2)
plt.plot(rnd1,rnd2, "r*")

plt.subplot(1,4,3)
plt.plot(carpim, dizi, "go--")

plt.subplot(1,4,4)
plt.plot(rnd2, rnd2, "#0f0f")
plt.show()
```
