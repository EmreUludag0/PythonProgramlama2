## Birden Fazla Sütün İle Çalışma


```python
rastgele = np.random.randint(10,20,(4,4)) #matris oluşturduk
print(len(rastgele))
#pd.DataFrame(rastgele, columns= ["baslik1","baslik2","baslik3"])
satirlar = []
kolonlar = []
for i in range(1,len(rastgele)+1):
    satirlar.append(i)
    kolonlar.append("baslik "+str(i))

myData = pd.DataFrame(rastgele,index=satirlar, columns=kolonlar)
myData
```

```python
print(type(myData)) #dataFrame'in tipini gösterir.
```
