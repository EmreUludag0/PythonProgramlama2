```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
yas = np.random.randint(20,40,5)
kilo = np.random.randint(25,50,5)
```

```python
plt.figure("ilk grafik", facecolor="orange", figsize=(8,8), dpi=100)
eksen = plt.axes() #graik içi
eksen.set_facecolor("#01ac88")
plt.plot(yas, kilo,"#fff", markersize = 7 , marker="*", label="YAS")
plt.plot(kilo,yas,"b:",markersize=7,marker="*", label="KİLO")
```
